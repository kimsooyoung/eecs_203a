```python
def read_raw_image(filename, height=480, width=640):
    with open(filename, 'rb') as f:
        data = np.frombuffer(f.read(), dtype=np.uint8)
    return data.reshape((height, width))
```

```python
L = 256
r = np.arange(L, dtype=np.float32)
gamma_values = [0.4, 2.5]
```

```python
def create_LUT(gamma):
    s = ((r/(L-1)) ** gamma) * (L-1)
    return np.clip(np.floor(s + 0.5), 0, L-1).astype(np.uint8)
```

```python
luts = {y: create_LUT(y) for y in gamma_values}
```

```python
plt.figure(figsize=(8,6))
for y, lut in luts.items():
    plt.plot(r, lut, label=f'y = {y}')
plt.title('GLT')
plt.xlabel('Input gray-level (r)')
plt.ylabel('Output gray-level (s)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
cat = read_raw_image('cat.raw')
```

```python
transformed = {}
for y, lut in luts.items():
    transformed[y] = lut[cat]

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(cat, cmap='gray', vmin=0, vmax=255)
plt.title('Original')
plt.axis('off')

for i, γ in enumerate(gamma_values, start=2):
    plt.subplot(1,3,i)
    plt.imshow(transformed[γ], cmap='gray', vmin=0, vmax=255)
    plt.title(f'γ = {γ}')
    plt.axis('off')

plt.tight_layout()
plt.show()

Image.fromarray(cat).save('cat_original.png')
for γ, img in transformed.items():
    Image.fromarray(img).save(f'cat_gamma_{γ}.png')
```
